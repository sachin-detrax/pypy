```python
from sklearn.datasets import load_wine
```

```python
df = load_wine()
```

```python
dir(df)
```

```python
inputs = df.data
```

```python
output = df.data
```

```python
inputs.shape
```

```python
output.shape
```

```python
df.feature_names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(inputs, columns = df.feature_names)
```

```python
df = pd.concat([df, pd.DataFrame(output)], axis=1)
```

```python
df
```

```python
X = df[["alcohol"]]
```

```python
X
```

```python
Y = df[["malic_acid"]]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
lr.predict(x_test)
```

```python
y_predict = lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```
